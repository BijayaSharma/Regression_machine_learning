# Bias 

Assumptions made by a model to make a function easier to learn. It is actually the error rate of the training data. When the error rate has a high value, we can call it high Bias and when the error rate has a low value, we call it low Bias.

Generally speaking, Bias is the difference between the actual value and predicted value.

# Variance 

The difference between the error rate of training data and testing data is called variance. If the difference is high then it is called high variance and when the difference of errors is low then it is called low variance. Usually we want to have a low variance for generalized our model.

If the machine learning model performs well with the training dataset, but doesn't perform well with the test dataset, then variance occurs.


# Overfitting

Overfitting happens when a model learns the detail and noise in the training data to the extent that it -vely impacts the performance of the model on new data. This means the noise or random fluctuations in the training data is picked up and learned as concepts by the model. The problem is that these concepts do not apply to new data and -vely impact the models ability to generalize. 


In overfitting case the model doesn't make accurate predictions on testing data. The cause of overfitting are the non-parametric and non-linear methods because these types of machine learning algorithms have more freedom in building the model based on the dataset and therefore they can really build unrealistic models. A solution to the avoid overfitting is using a lienar algorithm if we have linear data or using the parameters like the maximal depth if we are using decision trees.


The overfitted model has low bias and high variance


The chances of occurence of overfitting increases as much we provide training to our model. It means the more we train our model, the more chances of occuring the overfitted model.




# Reasons for overfitting 

1. High variance and low bias
2. The model too complex
3. The size of the training data

# Techniques to reduce overfitting

1. Increase training data
2. Reduce model complexity
3. Early stopping during the training phase (have an eye over the loss over the training period as soon as loss begins to increas stop training)
4. Ridge Regularization and Lasso Regularization
5. Use dropout for neural networks to tackle overfitting
6. Cross-Validation
7. Training with more data
8. Removing features


Cross validataion is a statistical method of evaluating and comparing learning algorithm by dividing data into 2 segments
One is used to learn or train a model and other used to validate the model

It is a technique for evaluating ML models by training several ML models on subsets of the available input data and evaluating them on the complementary subset of the data.

# Underfitting

It occurs when machine learning model is not able to capture the underlying trend of the data ( it only performs well on training data but performs poorly on testing data, just like trying to fit undersized pants!!). To avoid the overfitting in the model, the fed of training data can be stopped at an early stage, due to which the model may not learn enough from the training data. As a result, it may fail to find the best fit of the dominant trend in the data.

Its occurence means that our model or the algorithm doesn't fit the data well enough. It usually happens when we have fewer data to build an accurate model and also when we try to build a linear model with fewer non-linear data. In such cases, the rules of machine learning model are too easy and flexible to be applied such minimal data & therefore the model will probably make a lot of wrong predictions.

In underfitting, the model is not able to learn enough from the training data, and hence it reduces the accuracy and produces unreliable predictions.

A underfitted model has high bias and low variance.

# Reasons for Underfitting

1. High bias and low variance
2. The size of the training dataset used is not enough
3. The model is too simple
4. Training data is not cleaned and also contains noise in it.


# Techniques to reduce underfitting : 


1. Increase model complexity
2. Increase the number of features, performing feature engineering
3. Remove noise from the data
4. Increase duration of training to get better results

# Goodness of fit 

The "Goodness of Fit" term is taken from the statistics, the goal of machine learning models to achieve the goodness of fit. In statistics modeling, it defines how closely the result of predicted values match the true values of the dataset.

The model with good fit is between the underfitted and overfitted model, and ideally it makes predicted with 0 errors but in practice, it is difficult to achieve it.


When we train our model for a time, the errors in the training data go down, and same happens with test data. But if we train the model for long duration then the performance of the model may decrease due to the overfitting, as the mdoel also learn the noise present in the dataset. The errors in the test dataset start increasing, so the point, just before the raising of errors, is the good point and we can stop here for achieving a good model.

There are 2 other methods by which we can get a good point for our model, which are the resampling method to estimate model accuracy & validation dataset.
