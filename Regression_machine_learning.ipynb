# Regression 

It is a technique for investing the relationship between independent variables or features and a dependent variables or outcomes. (supervised learning)

Algorithms are trained to understand the relationship between independent variables & dependent variables. This model can be leveraged to predict the outcome of the new & unseen input data, or to fill a gap in the missing gap. 

THe approach to training data models requires labelled input and output training data. accurately labelled training data is vital.


Regression shows a line or curve that passes through all the datapoints on target-predictor graph in such a way that the vertical distance between the datapoints and the regression line is minimum. The distance between datapoints and line tells whether a model has captured a strong relationship or not.


## Regression analysis can also be used to fit the functions with 2 purposes :



1. We can estimate missing data within our data range (Interpolation)
2. We can estimate future data outside our data range (Extrapolation)

# Terminologies 

1. Dependent Variables : The main factor which we want to predict or understand. Also called target variable.

2. Independent variables : Factors which affect the dependent variables or which are used to predict the values of the dependent variables, also called predictor

3. Ouliers : contains low or high value

4. Multicollinearlity : If the independent variables are highly correlated with each other than other variables then this 
condition is called Multicollinearity. Shouldn't be present in the dataset because it creates problem while ranking the most affecting variables.



5. Underfitting & Overfitting : If algorithm works well with the training dataset but not with test dataset, then such problem is called Overfitting. 

        And if algorithm doesn't perform well even with the training dataset, then such problem is called Underfitting.

# Machine learning regression 

It is used as an approach to predict continuous outcomes in predictive modelling, so has utility in forecasting & predicting outcomes from data.

The model will learn the relationship between input & output data from labelling training data and then it can forecast future trends of predict outcomes from unseen input data, or be used to understand gaps in historic data.

As with all supervised machine learning, special care should be taken to ensure the labelled training data is representative of the overall population. If the training data is not representative, the predictive model will be overfit to data that doesn't represent new & unseen data. 

Machine learning regression model can be used to predict missing historic data, forecast sales in the business, also stock market.


# Common use of machine learning regression models

3. Creating time series visualisations

# 1. Linear Regression 

1. Statistical regression method which is used for predictive analysis.

2. shows the relationship between the continuous variables. (continuous variables means it can take an uncountable set of set of values or infinite set of values) 

3. If there is only one input variables then such linear regression is called simple linear regression & if there is more than one input variable, then such linear regression is called multiple linear regression. 

Y = AX + B is a linear regression equations. 
where Y is dependent (Target) variables while X is an Independent (predictor) variables

# 2. Logistic Regression 

1. This is used to solve the classification problems. In classification problems, we have dependent variables in a binary or discrete format such as 0 or 1.

2. Logistic Regression algorithm works with the categorical variables such as 0 or 1, Yes or No, True or False, Spam or not spam etc...

3. It is a predictive analysis alogrithm which works in the concept of probability

4. It is different from the linear regression algorithm in the term how they are used.

5. It uses sigmoid function or logistic function which is a complex cost function. This sigmoid function is used to model the data in logistic regression. 

The function is :: f(x) = 1/(1+exp-x), f(x) = output between 0 & 1, e is base of natural logarithm

6. Uses the concept of threshold levels, values above the threshold levels are rounded upto 1, & values below the threshold are rounded up to 0. 

7. There are 3 types of logistic regression : 
    
    1. Binary (0/1, pass/fail)
    2. Multi (cats, dogs, lions)
    3. Ordinal (low, medium, high)
    
    

# 3 Polynomial Regression 

1. It is a type of regression which models the non-linear dataset using a linear model 

2. Similar to the multiple linear regression, but it fits a non-linear curve between the values of x and corresponding conditional values of y.

3. Suppose there is a dataset which consists of datapoints which are present in a non-linear fashion, so for such case, linear regression won't best fit to those datapoints. So, polynomial regression is useful

4. In polynomial regression, the original features are transformed into polynomial features of given degree and then modeled using a linear model. Means that the datapoints are best fitted using a polynomial line.

5. The equation for polynomial regression also derived from linear regression equation that means linear regression equation Y = b0 + b1X is transformed into polynomial regression equation Y = b0+b1X + b2X**2 + b3X**3 +... + bnX**n


6. NOTE : 
       => This is different from Multiple linear regression in such a way that in Polynomial regression, a single element            has different degrees instead of multiple variables with the same degree.

#  4 Support Vector Regression 

Support vector machine is a supervised learning algorithm which can be used for regression as well as classification problems. So if we use it for regression problems then it is termed as Support Vector Regression. It works for continuous variables. 

Below are the keywords which are used in Support Vector Regression 

    1. Kernel : function used to map a lower dimensional data into higher dimensional data. 
    2. Hyperplane : In general SVM, it is a separation line between 2 classes, but in SVR, it is a line which helps to                          predict the continuous variables and cover most of the datapoints.
    
    3. Boundary line : These are the 2 lines apart from hyperplane, which creates a margin for datapoints.
    4. Support Vectors : These are the datapoints which are nearest to the hyperplane and opposite class.
    
   
In SVR, we always try to determine a hyperplane with a maximum margin, so that the maximum numbr of datapoints are covered in that margin. The main goal of SVR is to consider maximum datapoints within the boundary lines & the hyperplane (best-fit line) must contain maximum number of datapoints.
